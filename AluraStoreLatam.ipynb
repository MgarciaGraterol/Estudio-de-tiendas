### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

```python
# Agregamos columna de tienda
tienda1['tienda'] = 'tienda_1'
tienda2['tienda'] = 'tienda_2'
tienda3['tienda'] = 'tienda_3'
tienda4['tienda'] = 'tienda_4'

# Unir todos los datos
datos = pd.concat([tienda1, tienda2, tienda3, tienda4], ignore_index=True)

datos.head()
```

#1. Análisis de facturación



```python
datos['facturacion'] = datos['Precio'] * datos['Cantidad de cuotas']
facturacion_tienda = datos.groupby('tienda')['facturacion'].sum().sort_values(ascending=False)
facturacion_tienda
```

# 2. Ventas por categoría

```python
ventas_categoria = datos.groupby(['tienda', 'Categoría del Producto'])['Cantidad de cuotas'].sum().unstack()
ventas_categoria
```

# 3. Calificación promedio de la tienda


```python
calificaciones = datos.groupby('tienda')['Calificación'].mean().round(2)
calificaciones
```

# 4. Productos más y menos vendidos

```python
productos_mas_vendidos = datos.groupby('Producto')['Cantidad de cuotas'].sum().sort_values(ascending=False).head(5)
productos_menos_vendidos = datos.groupby('Producto')['Cantidad de cuotas'].sum().sort_values().head(5)

productos_mas_vendidos, productos_menos_vendidos
```

# 5. Envío promedio por tienda

```python
envio_promedio = datos.groupby('tienda')['Costo de envío'].mean().round(2)
envio_promedio
```
